# Question 2

```python
import pandas as pd
lst = ['A','B','C']
df = pd.DataFrame(columns=lst)
```

```python
df
```

```python
def reindex_df(df):
    def index_generator():
        i = 1
        while True:
            yield i
            i += 2
    
    new_index = index_generator()
    return df.set_index(pd.Series(new_index, index=df.index))
```

```python
reindex_df(df)
```

```python
df = pd.DataFrame(['siddesh.waje@gmail.com','shreyash@gmail.com','soham.waghule@gmail.com'],columns=['Email'])
```

```python
df
```

```python
def username_Creator(df):
    df['Username'] = df['Email'].str.split('@').str[0]
    return df
```

```python
username_Creator(df)
```

## The function uses the str.split() method to split each email address in the 'Email' column into two parts, separated by the '@' symbol. It then uses the .str[0] indexing operation to extract the first part, which is the username, and assigns the resulting values to a new column 'Username' using the assignment syntax df['Username'] =.
